```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Information below is my testing just using pandas .corr(); This wasn't part of Heather's example.
```

```python
dps_test = pd.read_csv("training_sample.csv")
dps_test.head()
```

```python
corr = dps_test.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
f, ax = plt.subplots(figsize=(11, 9))
```

```python
cmap = sn.diverging_palette(230, 20, as_cmap=True)
```

```python
# Draw the heatmap with the mask and correct aspect ratio
heatmap = sn.heatmap(corr, mask=mask, cmap=cmap,
           vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5
           ,xticklabels=True, yticklabels=True
#            , cbar_kws={"shrink": 1}
          )
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':15}, pad=12);
```

```python
dps_test_ordered_only = dps_test[(dps_test['ordered'] == 1)]
dps_test_ordered_only.head()
```
